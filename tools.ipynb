```python
import os


path = "./src/problems"
path_label = "'problems'" # / 'Algorithm.cpp',"
L=[]
def file_name(file_dir, label):   
    for root, dirs, files in os.walk(file_dir):        
        for dir in dirs:
            dirTemp = file_dir+'/'+dir
            labelTemp= label+" / '"+dir+"'"
            file_name(dirTemp,labelTemp)
        for file in files:
            if os.path.splitext(file)[1] == '.cpp':  # 想要保存的文件格式
                L.append(label+" / '"+file+"'")
            # for root, dirs, files in os.walk(dirTemp):
            #     for file in files:
            #         if os.path.splitext(file)[1] == 'cpp':  # 想要保存的文件格式
            #             L.append(dir+"/"+file)
file_name(path,path_label)
for l in L:
   print(l+",")
```
